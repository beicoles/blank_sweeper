```python
import impyute
import pandas as pd
import copy
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
from time import time

from sklearn.decomposition import MiniBatchDictionaryLearning
import matplotlib
import matplotlib.pyplot as plt
import scipy as sp
```

```python
imp = SimpleImputer(missing_values= np.nan, strategy='mean')

csv = open("Fireball_And_Bolide_Reports.csv", 'r')
all_data = csv.readlines()[1:]

for i in range(len(all_data)):
    all_data[i] = all_data[i][:-2].split(',')[1:]


nparray = np.array(all_data)[:,:3].tolist()

for arr in nparray:
    arr[0] = float(arr[0][:-1]) if arr[0][-1:] is 'N' else float(arr[0][:-1]) * -1.0
    
    arr[1] = float(arr[1][:-1]) if arr[1][-1:] is 'E' else float(arr[1][:-1]) * -1.0
train = []
target = []
for arr in nparray:
    if arr[2] is "":
        target.append(arr)
    else:
        train.append(arr)

print(np.array(train))
```

```python
arr =np.array(train).astype(np.float)
t0 = time()
dico = MiniBatchDictionaryLearning(n_components=100, alpha=1, n_iter=100)
V = dico.fit(arr).components_
dt = time() - t0
print('done in %.2fs.' % dt)
```

```python
target_np = np.array(target)
for npa in nparray:
    if npa[2] == "":
        npa[2]=np.nan 

result = impyute.fast_knn(np.array(nparray).astype(float))
#result = np.around(result,3)


target_arr = copy.deepcopy(nparray)
for npaa in target_arr:
    if npaa[2] is np.nan:
        npaa[2]=-1 

print(nparray)
type(nparray)

#print(result[np.lexsort((result[:,2],result[:,1],result[:,0]))])

```
